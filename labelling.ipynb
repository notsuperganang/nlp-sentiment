## 🧹 **Deduplication: Membersihkan Data dari Duplikasi!** 🔄📊  

Pada tahap ini, kita akan **menghapus duplikasi dalam dataset** menggunakan `pandas`. Dataset **`clean_data.csv`** mengandung teks yang mungkin memiliki **duplikasi**, sehingga kita perlu menyaring hanya **teks unik** untuk menjaga kualitas data.  

### 📌 **Mengapa Deduplication Itu Penting?**  
✅ **Mengurangi redundansi** dalam data  
✅ **Meningkatkan efisiensi** pemrosesan  
✅ **Meningkatkan akurasi model** dengan data yang lebih bersih  

### 🔍 **Bagaimana Cara Kerjanya?**  
1️⃣ **Memuat dataset** dengan `pandas`  
2️⃣ **Mengidentifikasi duplikasi** berdasarkan kolom `extracted_text`  
3️⃣ **Menghitung jumlah duplikasi** untuk memahami skala masalah  
4️⃣ **Menghapus duplikasi** dengan memilih teks yang memiliki **engagement (likes + reply_count) tertinggi**  
5️⃣ **Menyimpan dataset yang telah dibersihkan** ke **`clean_data_deduplicated.csv`**  

### 📊 **Ringkasan Proses Deduplication**  
📂 **Dataset awal:** 16.555 baris  
🔄 **Duplikasi ditemukan:** 2770 baris  
✅ **Dataset akhir setelah deduplication:** 13785 baris  

````python
import pandas as pd
import os

# Input and output file paths
INPUT_CSV = 'full-data/result/clean_data.csv'
OUTPUT_CSV = 'full-data/result/clean_data_deduplicated.csv'

def deduplicate_dataset():
    print("📊 Loading dataset...")
    # Read the original dataset
    df = pd.read_csv(INPUT_CSV)
    total_records = len(df)
    print(f"✅ Dataset loaded. Total records: {total_records}")
    
    # Count unique texts
    unique_texts_count = df['extracted_text'].nunique()
    print(f"📝 Unique texts in dataset: {unique_texts_count}")
    print(f"🔄 Duplicates: {total_records - unique_texts_count} records")
    
    # Identify duplicates
    duplicate_mask = df.duplicated(subset=['extracted_text'], keep=False)
    duplicates_df = df[duplicate_mask]
    print(f"🔍 Found {len(duplicates_df)} records with duplicate text")
    
    # Sample of duplicates (optional, for inspection)
    if len(duplicates_df) > 0:
        print("\n📋 Sample of duplicate texts:")
        # Group by text to show examples of duplicates
        for i, (text, group) in enumerate(duplicates_df.groupby('extracted_text')):
            if i >= 3:  # Show just 3 examples
                break
            print(f"  Text #{i+1}: '{text[:50]}{'...' if len(text) > 50 else ''}'")
            print(f"  Appears {len(group)} times")
            print(f"  Sample record: {group.iloc[0][['likes', 'reply_count']].to_dict()}")
            print()
    
    # Perform deduplication
    print("\n🧹 Removing duplicates...")
    
    # Create a temporary metric for engagement
    df['engagement'] = df['likes'].fillna(0) + df['reply_count'].fillna(0)
    
    # Sort by engagement in descending order, then deduplicate
    df = df.sort_values('engagement', ascending=False)
    deduplicated_df = df.drop_duplicates(subset=['extracted_text'], keep='first')
    
    # Remove the temporary engagement column
    deduplicated_df = deduplicated_df.drop('engagement', axis=1)
    
    print(f"✅ Deduplication complete. Records remaining: {len(deduplicated_df)}")
    print(f"🗑️ Removed {total_records - len(deduplicated_df)} duplicate records")
    
    # Save deduplicated dataset
    print(f"💾 Saving deduplicated dataset to {OUTPUT_CSV}...")
    deduplicated_df.to_csv(OUTPUT_CSV, index=False)
    print("✅ Deduplicated dataset saved successfully!")

if __name__ == "__main__":
    # Make sure output directory exists
    os.makedirs(os.path.dirname(OUTPUT_CSV), exist_ok=True)
    deduplicate_dataset()
    print("✨ Process completed!")
````

## ✂️ **Membagi Dataset Menjadi 6 Bagian** 🏗️📊  

Dalam tahap ini, kita akan **membagi dataset yang telah dide-duplikasi** menjadi **6 bagian** menggunakan `pandas`. Tujuan utama dari proses ini adalah **mengurangi risiko crash** saat melakukan **labelling otomatis menggunakan API LLM**.  

### 🔍 **Mengapa Perlu Dibagi Menjadi 6 Bagian?**  
✅ **Mencegah crash** akibat pemrosesan data yang terlalu besar dalam satu waktu  
✅ **Mempercepat eksekusi** dengan membagi beban kerja menjadi batch yang lebih kecil  
✅ **Memudahkan troubleshooting** jika terjadi error saat labelling  
✅ **Memastikan distribusi data yang merata** agar proses labelling lebih optimal  

### 📌 **Bagaimana Cara Kerjanya?**  
1️⃣ **Memuat dataset deduplikasi** dari `clean_data_deduplicated.csv`  
2️⃣ **Menghitung ukuran setiap bagian**, dengan menyesuaikan sisa data agar tidak ada kehilangan baris  
3️⃣ **Mengacak urutan data** untuk memastikan distribusi yang merata  
4️⃣ **Membagi dataset menjadi 6 bagian** dengan jumlah baris yang seimbang  
5️⃣ **Menyimpan setiap bagian ke file CSV terpisah** di dalam folder `chunks/`  

### 📊 **Ringkasan Proses Pembagian Dataset**  
📂 **Dataset awal:** 13785 baris  
📦 **Jumlah chunk:** 6  
🔄 **Ukuran tiap chunk:** Sebagian besar memiliki `13785/6` baris, dengan sedikit penyesuaian agar merata  
📁 **Lokasi hasil:** `full-data/result/chunks/`  

Dengan dataset yang telah dibagi menjadi **6 bagian**, kita siap melanjutkan ke tahap **labelling otomatis menggunakan LLM** dengan risiko crash yang lebih rendah! 🚀🔥

````python
import pandas as pd
import os
import math
import numpy as np

# Configuration
INPUT_CSV = 'full-data/result/clean_data_deduplicated.csv'
OUTPUT_DIR = 'full-data/result/chunks/'
NUM_CHUNKS = 6
RANDOM_SEED = 42  # For reproducibility if you want to shuffle the data

def split_dataset():
    # Create output directory if it doesn't exist
    os.makedirs(OUTPUT_DIR, exist_ok=True)
    
    print("📊 Loading deduplicated dataset...")
    df = pd.read_csv(INPUT_CSV)
    total_records = len(df)
    print(f"✅ Dataset loaded successfully. Total unique records: {total_records}")
    
    # Calculate chunk size and distribution
    base_chunk_size = math.floor(total_records / NUM_CHUNKS)
    remainder = total_records % NUM_CHUNKS
    
    print(f"\n📦 Splitting dataset into {NUM_CHUNKS} chunks:")
    print(f"  - Base chunk size: {base_chunk_size} records per chunk")
    if remainder > 0:
        print(f"  - First {remainder} chunks will have {base_chunk_size + 1} records")
    
    # Optional: shuffle dataset to ensure random distribution
    # Comment this out if you want to maintain original order
    df = df.sample(frac=1, random_state=RANDOM_SEED).reset_index(drop=True)
    print("🔀 Shuffled dataset to ensure random distribution of records")
    
    # Split and save chunks
    start_idx = 0
    for i in range(NUM_CHUNKS):
        # Calculate size of current chunk (add 1 to first 'remainder' chunks)
        current_chunk_size = base_chunk_size + (1 if i < remainder else 0)
        end_idx = start_idx + current_chunk_size
        
        # Extract chunk
        chunk_df = df.iloc[start_idx:end_idx].copy()
        
        # Generate output filename
        output_file = os.path.join(OUTPUT_DIR, f"chunk_{i+1}_of_{NUM_CHUNKS}.csv")
        
        # Save chunk
        chunk_df.to_csv(output_file, index=False)
        
        # Print statistics for this chunk
        print(f"\n✅ Chunk {i+1}/{NUM_CHUNKS} created:")
        print(f"  - Records: {len(chunk_df)}")
        print(f"  - Range: rows {start_idx+1} to {end_idx}")
        print(f"  - Saved to: {output_file}")
        
        # Update start index for next chunk
        start_idx = end_idx
    
    print("\n🎉 Dataset splitting completed successfully!")
    print(f"📁 All chunks saved to: {OUTPUT_DIR}")
    print("\nTo process these chunks, update your sentiment labeling script with:")
    print("INPUT_CSV = '[path to chunk file]'")

if __name__ == "__main__":
    split_dataset()
````

    print(f"MEMULAI PROSES UNTUK CHUNK {i}/6")
    print(f"{'='*60}\n")
    
    # Jalankan proses labeling untuk chunk ini
    main()
    
    print(f"\n{'='*60}")

Dengan pendekatan ini, sistem dapat memproses dataset skala besar secara bertahap dan dengan penggunaan memori yang terkontrol.

````python
import pandas as pd
import numpy as np
import os
import csv
import random
import time
from tqdm import tqdm
from g4f.client import Client

# Konfigurasi
INPUT_CSV = 'full-data/result/chunks/chunk_1_of_6.csv'  # Data yang telah dipreprocessing
PROGRESS_CSV = 'label-data/progress_sentiment_1_to_6.csv'  # File untuk menyimpan progres
OUTPUT_CSV = 'label-data/labeled_sentiment_1_to_6.csv'  # File output hasil labeling
BATCH_SIZE = 20  # Jumlah teks yang dilabel per batch
MAX_RETRIES = 5  # Batas percobaan ulang jika terjadi error
DELAY_BETWEEN_RETRIES = 10  # Waktu jeda antara percobaan dalam detik
MAX_PROCESSING_TIME = 7200  # Batas waktu proses (dalam detik, default 2 jam)

# Pastikan direktori output ada
os.makedirs('label-data', exist_ok=True)

# Inisialisasi client g4f
client = Client()

# Template Prompt
PROMPT_TEMPLATE = """
Klasifikasikan {batch_size} komentar TikTok berikut mengenai topik "#kaburajadulu" berdasarkan sentimen:

- POSITIF: Mendukung, menyetujui, atau berempati dengan konsep #KaburAjaDulu. Komentar yang mengekspresikan keinginan untuk pindah ke luar negeri.
- NEGATIF: Menolak atau tidak setuju dengan konsep #KaburAjaDulu. Komentar yang mendorong untuk tetap tinggal di Indonesia.
- NETRAL: Tidak jelas menunjukkan dukungan atau penolakan terhadap konsep #KaburAjaDulu. Komentar yang hanya informatif.

INSTRUKSI PENTING:
1. Berikan TEPAT {batch_size} label (tidak boleh kurang atau lebih)
2. Gunakan hanya: positif, negatif, atau netral
3. Format: positif,negatif,netral,... (tanpa spasi setelah koma)

Teks ({batch_size} buah):
{text_list}

Jawaban (harus tepat {batch_size} label):"""

# Template Prompt untuk percobaan ulang (lebih ketat)
RETRY_PROMPT_TEMPLATE = """
SANGAT PENTING! Klasifikasikan TEPAT {batch_size} komentar TikTok berikut terkait #KaburAjaDulu:

INSTRUKSI YANG WAJIB DIIKUTI:
- Berikan TEPAT {batch_size} label (tidak boleh {batch_size_minus} atau {batch_size_plus})
- Format: positif,negatif,netral,... (tanpa spasi setelah koma)
- Hanya gunakan kata-kata berikut: positif, negatif, netral

PEDOMAN SENTIMEN #KABURAJADULU:
- POSITIF = Mendukung keinginan meninggalkan Indonesia, membandingkan kondisi lebih baik di luar negeri
- NEGATIF = Menentang ide pergi dari Indonesia, menyarankan bertahan, kritik terhadap yang ingin pergi
- NETRAL = Tidak jelas mendukung/menolak, hanya menyatakan fakta

Teks ({batch_size} buah):
{text_list}

JAWABAN (HARUS TEPAT {batch_size} LABEL):"""

def check_for_duplicates(df):
    """Check for duplicate extracted_text values and report them"""
    # Count occurrences of each extracted_text value
    text_counts = df['extracted_text'].value_counts()
    
    # Find values that appear more than once
    duplicates = text_counts[text_counts > 1]
    
    if len(duplicates) > 0:
        print(f"⚠️ PERINGATAN: Ditemukan {len(duplicates)} teks unik yang memiliki duplikat")
        print(f"⚠️ Total {duplicates.sum()} record dengan teks duplikat")
        
        # Show some examples of duplicates
        print(f"📝 Contoh teks dengan duplikat:")
        count = 0
        for text, frequency in duplicates.items():
            if count >= 3:  # Show at most 3 examples
                break
            print(f"  {count+1}. '{text[:50]}...' muncul {frequency} kali")
            count += 1
    
    return len(duplicates) > 0

def deduplicate_progress_file():
    """Handles duplication in the progress file to ensure consistent labels"""
    if not os.path.exists(PROGRESS_CSV):
        return False
        
    try:
        # Read progress file
        df = pd.read_csv(PROGRESS_CSV)
        original_count = len(df)
        
        # Check if we have duplicates
        has_dups = check_for_duplicates(df)
        if not has_dups:
            return False
            
        print("🔄 Menyelaraskan label untuk teks yang sama...")
        
        # Get all labeled texts with their labels
        labeled_df = df[df['sentiment_label'].notna()].copy()
        labeled_texts = {}
        
        # For each labeled text, store the most common label
        for text, group in labeled_df.groupby('extracted_text'):
            # Get the most common label for this text
            label_counts = group['sentiment_label'].value_counts()
            most_common_label = label_counts.index[0]
            labeled_texts[text] = most_common_label
            
            # If there are conflicting labels, report it
            if len(label_counts) > 1:
                print(f"⚠️ Teks '{text[:30]}...' memiliki {len(label_counts)} label berbeda. Menggunakan '{most_common_label}'")
        
        # Apply consistent labels to all duplicates
        for idx, row in df.iterrows():
            text = row['extracted_text']
            if text in labeled_texts:
                df.at[idx, 'sentiment_label'] = labeled_texts[text]
        
        # Save the updated progress file
        df.to_csv(PROGRESS_CSV, index=False, quoting=csv.QUOTE_ALL)
        
        # Report how many records were updated
        final_labeled = df['sentiment_label'].notna().sum()
        print(f"✅ Penyelarasan selesai. Total berlabel sekarang: {final_labeled}")
        
        return True
        
    except Exception as e:
        print(f"❌ Error saat menyelaraskan label: {str(e)}")
        return False

def load_data():
    """Load data and handle progress tracking properly"""
    try:
        # Memuat dataset utama terlebih dahulu
        print(f"📊 Memuat dataset dari {INPUT_CSV}...")
        main_df = pd.read_csv(INPUT_CSV)
        total_records = len(main_df)
        print(f"✅ Dataset berhasil dimuat. Total record: {total_records}")
        
        # Check for duplicates in main dataset
        has_duplicates = check_for_duplicates(main_df)
        
        # Cek apakah file progress ada
        if os.path.exists(PROGRESS_CSV):
            print(f"🔍 File progress ditemukan: {PROGRESS_CSV}")
            
            # Fix duplicates in progress file if any
            if has_duplicates:
                deduplicate_progress_file()
                
            try:
                progress_df = pd.read_csv(PROGRESS_CSV)
                progress_count = len(progress_df)
                labeled_count = progress_df['sentiment_label'].notna().sum()
                
                print(f"📈 Status progress: {labeled_count}/{progress_count} record telah dilabel")
                
                # Verifikasi bahwa progress file sesuai dengan dataset utama
                if progress_count != total_records:
                    print(f"⚠️ Jumlah record di file progress ({progress_count}) tidak sama dengan dataset utama ({total_records})")
                    print("🔄 Membuat file progress baru dengan dataset lengkap...")
                    
                    # Membuat progress baru dengan mempertahankan label yang sudah ada
                    main_df['sentiment_label'] = None
                    
                    # Membuat dictionary dari teks ke label dari progress file
                    text_to_label = {}
                    for idx, row in progress_df.iterrows():
                        if pd.notna(row.get('sentiment_label')):
                            text_to_label[row['extracted_text']] = row['sentiment_label']
                    
                    # Salin label yang sudah ada ke dataset utama berdasarkan extracted_text
                    update_count = 0
                    for idx, row in main_df.iterrows():
                        text = row['extracted_text']
                        if text in text_to_label:
                            main_df.at[idx, 'sentiment_label'] = text_to_label[text]
                            update_count += 1
                    
                    # Simpan progress baru
                    main_df.to_csv(PROGRESS_CSV, index=False, quoting=csv.QUOTE_ALL)
                    print(f"✅ File progress baru dibuat dengan {update_count} label yang dipertahankan")
                    return main_df.to_dict('records')
                
                return progress_df.to_dict('records')
            
            except Exception as e:
                print(f"❌ Error membaca file progress: {str(e)}")
                print("🔄 Membuat file progress baru...")
                main_df['sentiment_label'] = None
                main_df.to_csv(PROGRESS_CSV, index=False, quoting=csv.QUOTE_ALL)
                return main_df.to_dict('records')
        
        # Jika tidak ada file progress, buat baru
        print("🆕 Membuat file progress baru...")
        main_df['sentiment_label'] = None
        main_df.to_csv(PROGRESS_CSV, index=False, quoting=csv.QUOTE_ALL)
        return main_df.to_dict('records')
        
    except Exception as e:
        print(f"❌ Error fatal saat memuat data: {str(e)}")
        exit(1)

def create_batches(data):
    """Create batches of unlabeled data with duplicate prevention"""
    # First, identify unique texts that need labeling
    unlabeled_texts = set()
    text_to_records = {}
    
    for record in data:
        text = record.get('extracted_text', '')
        if pd.isna(record.get('sentiment_label')):
            unlabeled_texts.add(text)
            
            # Group records by text to handle duplicates
            if text not in text_to_records:
                text_to_records[text] = []
            text_to_records[text].append(record)
    
    # Create a list of unique unlabeled texts
    unique_unlabeled = list(unlabeled_texts)
    total_unique = len(unique_unlabeled)
    total_records = len(data)
    total_unlabeled_records = len([d for d in data if pd.isna(d.get('sentiment_label'))])
    
    print(f"📋 Statistik data yang belum dilabel:")
    print(f"  - Total record belum berlabel: {total_unlabeled_records}/{total_records}")
    print(f"  - Teks unik belum berlabel: {total_unique}")
    
    if total_unique == 0:
        print("🎉 Semua teks unik telah dilabel! Tidak ada yang perlu diproses.")
        return []
    
    # Shuffle untuk randomisasi
    random.shuffle(unique_unlabeled)
    
    # Buat batch dengan ukuran yang tepat (satu record per teks unik)
    batches = []
    for i in range(0, total_unique, BATCH_SIZE):
        batch_texts = unique_unlabeled[i:i+BATCH_SIZE]
        # Take one record per unique text
        current_batch = [text_to_records[text][0] for text in batch_texts]
        batches.append(current_batch)
    
    print(f"🧮 Membagi {total_unique} teks unik menjadi {len(batches)} batch")
    return batches

def generate_prompt(batch):
    """Generate prompt for the current batch"""
    # Pastikan setiap teks memiliki identifier yang unik (indeks + hash ringkas)
    texts = []
    for idx, item in enumerate(batch):
        text = item['extracted_text']
        if not isinstance(text, str) or text.strip() == '':
            text = "(komentar kosong)"
        texts.append(f"{idx+1}. {text}")
    
    return PROMPT_TEMPLATE.format(
        batch_size=len(batch),
        text_list='\n'.join(texts)
    )

def parse_response(response, batch):
    """Parse model response and handle potential issues"""
    valid_labels = ['positif', 'negatif', 'netral']
    
    try:
        # Clean up the response
        cleaned_response = response.strip()
        
        # Handle multiple potential response formats
        if ',' in cleaned_response:
            # Looks like a comma-separated list
            raw_labels = []
            for label in cleaned_response.split(','):
                label = label.strip().lower()
                if label in valid_labels:
                    raw_labels.append(label)
        else:
            # Try to find a list in the response
            lines = cleaned_response.split('\n')
            for line in lines:
                if all(label in line.lower() for label in valid_labels):
                    # Found a line with all labels
                    raw_labels = []
                    for word in line.lower().split():
                        word = word.strip('.,;:()[]{}"\'-')
                        if word in valid_labels:
                            raw_labels.append(word)
                    if raw_labels:
                        break
            else:
                # No clear list found, try one more method
                raw_labels = []
                for word in cleaned_response.lower().split():
                    word = word.strip('.,;:()[]{}"\'-')
                    if word in valid_labels:
                        raw_labels.append(word)
        
        # Verify we have the right number of labels
        if len(raw_labels) != len(batch):
            print(f"⚠️ Jumlah label ({len(raw_labels)}) tidak sesuai dengan jumlah teks ({len(batch)})")
            
            # Try to fix by extending or truncating
            if len(raw_labels) > len(batch):
                raw_labels = raw_labels[:len(batch)]
                print(f"✂️ Memotong ke {len(batch)} label")
            elif len(raw_labels) < len(batch) and raw_labels:
                # Fill missing labels with the most common label
                most_common = max(set(raw_labels), key=raw_labels.count)
                while len(raw_labels) < len(batch):
                    raw_labels.append(most_common)
                print(f"➕ Menambah hingga {len(batch)} label dengan '{most_common}'")
            else:
                # Can't fix - not enough data
                raise ValueError("Tidak dapat mengekstrak label yang cukup")
        
        # Apply labels to batch
        for i, label in enumerate(raw_labels):
            batch[i]['sentiment_label'] = label
        
        return batch
    
    except Exception as e:
        print(f"❌ Error saat parsing respons: {str(e)}")
        print(f"💬 Respons yang tidak valid: {response}")
        return None

def save_progress(data):
    """Save progress with verification"""
    try:
        df = pd.DataFrame(data)
        
        # Verifikasi sebelum menyimpan
        total_records = len(df)
        labeled_count = df['sentiment_label'].notna().sum()
        
        # Tampilkan informasi verifikasi
        print(f"📊 Verifikasi sebelum menyimpan:")
        print(f"  - Total record: {total_records}")
        print(f"  - Total berlabel: {labeled_count} ({labeled_count/total_records*100:.1f}%)")
        print(f"  - Total belum berlabel: {total_records - labeled_count}")
        
        # Check for any inconsistencies in labeling of duplicate texts
        text_to_labels = {}
        inconsistent_texts = []
        
        for idx, row in df.iterrows():
            text = row['extracted_text']
            label = row.get('sentiment_label')
            
            if pd.notna(label):
                if text not in text_to_labels:
                    text_to_labels[text] = label
                elif text_to_labels[text] != label:
                    inconsistent_texts.append(text)
        
        if inconsistent_texts:
            print(f"⚠️ Ditemukan {len(inconsistent_texts)} teks dengan label tidak konsisten")
            print("🔄 Menyeragamkan label untuk teks yang sama...")
            
            # Fix inconsistencies
            for text in inconsistent_texts:
                # Find all occurrences of this text
                matching_indices = df[df['extracted_text'] == text].index
                # Get all labels for this text
                labels = df.loc[matching_indices, 'sentiment_label'].dropna()
                # Find most common label
                most_common = labels.value_counts().idxmax()
                # Apply consistent label
                df.loc[matching_indices, 'sentiment_label'] = most_common
            
            print(f"✅ Label telah diseragamkan")
            labeled_count = df['sentiment_label'].notna().sum()
            print(f"📊 Setelah penyeragaman: {labeled_count} record berlabel")
        
        # Simpan dengan backup
        backup_file = f"{PROGRESS_CSV}.bak"
        if os.path.exists(PROGRESS_CSV):
            os.replace(PROGRESS_CSV, backup_file)
            print(f"💾 Backup dibuat: {backup_file}")
        
        df.to_csv(PROGRESS_CSV, index=False, quoting=csv.QUOTE_ALL)
        print(f"💾 Progress tersimpan: {labeled_count}/{total_records} ({labeled_count/total_records*100:.1f}%)")
        
        return True
    except Exception as e:
        print(f"❌ Error saat menyimpan progress: {str(e)}")
        print("⚠️ Mencoba memulihkan dari backup...")
        try:
            if os.path.exists(f"{PROGRESS_CSV}.bak"):
                os.replace(f"{PROGRESS_CSV}.bak", PROGRESS_CSV)
                print("✅ Pemulihan dari backup berhasil")
        except:
            print("❌ Gagal memulihkan dari backup")
        return False

def label_batch(batch):
    """Label a batch of texts with robust error handling and retry logic"""
    # Tunjukkan contoh teks yang akan dilabel (hanya beberapa)
    print(f"\n🔍 Contoh teks yang akan dilabel:")
    for i, item in enumerate(batch[:3]):  # Tampilkan maksimal 3
        text = item['extracted_text']
        if not isinstance(text, str):
            text = str(text)
        print(f"  {i+1}. {text[:50]}{'...' if len(text) > 50 else ''}")
    if len(batch) > 3:
        print(f"  ... dan {len(batch) - 3} lainnya")
    
    # Coba dengan tingkat ketegasan yang meningkat
    for attempt in range(MAX_RETRIES):
        try:
            # Gunakan prompt yang lebih tegas untuk percobaan ulang
            if attempt == 0:
                prompt = generate_prompt(batch)
            else:
                prompt = RETRY_PROMPT_TEMPLATE.format(
                    batch_size=len(batch),
                    batch_size_minus=len(batch)-1,
                    batch_size_plus=len(batch)+1,
                    text_list='\n'.join([f"{idx+1}. {item['extracted_text']}" for idx, item in enumerate(batch)])
                )
            
            # Reduce temperature on retries (more deterministic)
            temperature = max(0.1, 0.2 - attempt * 0.05)
            
            print(f"🤖 Memanggil API (percobaan {attempt+1}/{MAX_RETRIES}, temp={temperature:.2f})...")
            
            response = client.chat.completions.create(
                model="gpt-4o-mini",
                messages=[{"role": "user", "content": prompt}],
                temperature=temperature
            )
            
            content = response.choices[0].message.content
            print(f"✅ Respons diterima ({len(content)} karakter)")
            
            # Coba parsing respons
            result = parse_response(content, batch)
            
            if result:
                # Tampilkan distribusi label
                labels = [item['sentiment_label'] for item in result]
                label_counts = {label: labels.count(label) for label in set(labels)}
                print("🏷️ Label terdeteksi:")
                for lbl, count in sorted(label_counts.items(), key=lambda x: -x[1]):
                    print(f"  - {lbl}: {count}x ({count/len(labels)*100:.1f}%)")
                
                return result
            else:
                print(f"⚠️ Gagal mengekstrak label, mencoba lagi...")
            
        except Exception as e:
            print(f"⚠️ Percobaan {attempt+1}/{MAX_RETRIES} gagal: {str(e)}")
        
        # Tunggu sebelum mencoba lagi, dengan waktu yang semakin lama
        retry_delay = DELAY_BETWEEN_RETRIES * (attempt + 1)
        print(f"⏳ Menunggu {retry_delay} detik sebelum mencoba lagi...")
        time.sleep(retry_delay)
    
    print("❌ Semua percobaan gagal untuk batch ini")
    return None

def update_original_data(data, labeled_batch):
    """Update original data with new labels, ensuring ALL matching texts get the same label"""
    text_to_label = {}
    
    # First, collect all the labeled texts and their labels
    for item in labeled_batch:
        text = item['extracted_text']
        label = item['sentiment_label']
        
        if not isinstance(text, str) or not text or pd.isna(label):
            continue
        
        text_to_label[text] = label
    
    # Count how many records will be affected
    affected_count = sum(1 for original in data 
                         if original['extracted_text'] in text_to_label 
                         and pd.isna(original.get('sentiment_label')))
    
    # Now update ALL matching records in the original data
    for original in data:
        text = original['extracted_text']
        if text in text_to_label:
            original['sentiment_label'] = text_to_label[text]
    
    # Report how many texts and total records were updated
    print(f"✅ {len(text_to_label)} teks unik diterapkan ke {affected_count} record yang belum berlabel")
    return affected_count

def main():
    print("\n" + "="*50)
    print("🚀 MEMULAI PROSES LABELING SENTIMENT #KABURAJADULU")
    print("="*50)
    
    # Track processing time
    start_time = time.time()
    processing_count = 0
    
    # Load data
    data = load_data()
    total_records = len(data)
    total_unlabeled = len([d for d in data if pd.isna(d.get('sentiment_label'))])
    
    # Count unique texts
    unique_texts = len(set(d['extracted_text'] for d in data))
    unique_unlabeled = len(set(d['extracted_text'] for d in data if pd.isna(d.get('sentiment_label'))))
    
    print(f"\n📊 STATISTIK DATASET:")
    print(f"  - Total record: {total_records}")
    print(f"  - Total teks unik: {unique_texts}")
    print(f"  - Record berlabel: {total_records - total_unlabeled} ({(total_records - total_unlabeled)/total_records*100:.1f}%)")
    print(f"  - Record belum berlabel: {total_unlabeled} ({total_unlabeled/total_records*100:.1f}%)")
    print(f"  - Teks unik belum berlabel: {unique_unlabeled}")
    
    # Create batches for processing
    batches = create_batches(data)
    
    # Jika semua data sudah berlabel, langsung ke output
    if not batches:
        print("\n✅ Semua data sudah memiliki label. Langsung ke output...")
    else:
        # Process batches with progress tracking
        with tqdm(total=unique_unlabeled, desc="🚀 Progress Labeling", unit="teks") as pbar:
            for batch_idx, batch in enumerate(batches):
                # Check if we've exceeded the time limit
                elapsed_time = time.time() - start_time
                if elapsed_time > MAX_PROCESSING_TIME:
                    print(f"\n⏱️ Batas waktu pemrosesan ({MAX_PROCESSING_TIME/3600:.1f} jam) tercapai.")
                    print(f"📈 Telah memproses {processing_count} dari {unique_unlabeled} teks unik.")
                    break
                
                # Print batch info
                print(f"\n\n{'='*50}")
                print(f"🔖 BATCH {batch_idx+1}/{len(batches)} ({len(batch)} teks unik)")
                print(f"{'='*50}")
                
                # Process the batch
                labeled_batch = label_batch(batch)
                
                if labeled_batch:
                    # Update original data - all copies of each text
                    update_original_data(data, labeled_batch)
                    processing_count += len(labeled_batch)  # Count unique texts, not records
                    
                    # Save progress regularly
                    save_progress(data)
                    
                    # Update progress bar based on unique texts processed
                    pbar.update(len(labeled_batch))
                else:
                    print(f"⚠️ Batch {batch_idx+1} gagal diproses. Melanjutkan ke batch berikutnya...")
    
    # Calculate final statistics
    final_labeled = len([d for d in data if not pd.isna(d.get('sentiment_label'))])
    final_unique_labeled = len(set(d['extracted_text'] for d in data if not pd.isna(d.get('sentiment_label'))))
    
    print(f"\n📊 STATISTIK AKHIR:")
    print(f"  - Total record: {total_records}")
    print(f"  - Total teks unik: {unique_texts}")
    print(f"  - Record berlabel: {final_labeled} ({final_labeled/total_records*100:.1f}%)")
    print(f"  - Record belum berlabel: {total_records - final_labeled} ({(total_records - final_labeled)/total_records*100:.1f}%)")
    print(f"  - Teks unik berlabel: {final_unique_labeled} ({final_unique_labeled/unique_texts*100:.1f}%)")
    
    # Save final output with relevant columns
    print("\n💾 Menyimpan hasil akhir...")
    
    # Define columns to include in output, in case some are missing
    desired_columns = ['text', 'extracted_text', 'cleaned_text', 'normalized_text', 
                       'processed', 'sentiment_label', 'likes', 'reply_count', 
                       'created_time_formatted']
    
    # Get available columns
    available_columns = list(data[0].keys())
    output_columns = [col for col in desired_columns if col in available_columns]
    
    # Create output dataframe
    output_df = pd.DataFrame(data)[output_columns]
    
    # Save output
    output_df.to_csv(OUTPUT_CSV, index=False, quoting=csv.QUOTE_ALL)
    print(f"🎉 PROSES SELESAI! Hasil disimpan di {OUTPUT_CSV}")
    
    # Print execution summary
    elapsed_time = time.time() - start_time
    print(f"\n⏱️ Total waktu eksekusi: {elapsed_time/60:.1f} menit")
    print(f"📊 Record berlabel: {final_labeled}/{total_records} ({final_labeled/total_records*100:.1f}%)")
    print(f"📊 Teks unik berlabel: {final_unique_labeled}/{unique_texts} ({final_unique_labeled/unique_texts*100:.1f}%)")

if __name__ == "__main__":
    try:
        main()
    except KeyboardInterrupt:
        print("\n\n⚠️ Proses dihentikan oleh pengguna.")
        print("💾 Menyimpan progress terakhir...")
        # Would save progress here if we had access to the data variable
        print("👋 Program berakhir.")
    except Exception as e:
        print(f"\n\n❌ ERROR FATAL: {str(e)}")
        print("👋 Program berakhir dengan kesalahan.")
````

## 🔄 **Penjelasan Kode Penggabungan File Sentimen**

### ⚙️ **Fungsi Utama: `merge_sentiment_files()`**

Kode ini berfungsi untuk menggabungkan beberapa file CSV hasil labeling sentimen menjadi satu file tunggal. Sederhananya, kode melakukan proses:

1. **Identifikasi File** - Mendaftarkan semua file CSV hasil labeling yang perlu digabungkan
2. **Pembacaan Data** - Membaca setiap file CSV dan menyimpannya sebagai DataFrame
3. **Penggabungan** - Menggabungkan semua DataFrame menjadi satu dataset tunggal
4. **Analisis** - Menyajikan statistik dari dataset gabungan
5. **Penyimpanan** - Menyimpan hasil penggabungan ke file CSV baru

### 📈 **Output yang Dihasilkan**

Hasil akhir dari kode ini adalah:

1. **File CSV Gabungan** - Dataset tunggal yang berisi semua record dari seluruh chunk
2. **Laporan Statistik** - Tampilan distribusi label sentimen dari keseluruhan dataset
3. **Informasi Validasi** - Peringatan tentang adanya duplikat jika terdeteksi

Dengan dataset gabungan ini, analisis sentimen komprehensif terhadap topik "#KaburAjaDulu" dapat dilakukan secara menyeluruh.

````python
# Directory where your labeled files are stored
input_dir = 'label-data/'
# Output file path
output_file = 'label-data/labeled_sentiment_combined.csv'

def merge_sentiment_files():
    print("🔄 Starting the merging process...")
    
    # List all the CSV files to merge
    files_to_merge = [
        'labeled_sentiment_1_to_6.csv',
        'labeled_sentiment_2_of_6.csv',
        'labeled_sentiment_3_of_6.csv',
        'labeled_sentiment_4_of_6.csv',
        'labeled_sentiment_5_of_6.csv',
        'labeled_sentiment_6_of_6.csv'
    ]
    
    # Create an empty list to store DataFrames
    dataframes = []
    total_records = 0
    
    # Read each file and append to the list
    for file_name in files_to_merge:
        file_path = os.path.join(input_dir, file_name)
        
        # Check if file exists
        if not os.path.exists(file_path):
            print(f"⚠️ Warning: File {file_path} not found. Skipping.")
            continue
        
        print(f"📄 Reading file: {file_name}")
        df = pd.read_csv(file_path)
        
        # Print statistics for this file
        records = len(df)
        labeled = df['sentiment_label'].notna().sum()
        
        print(f"  - Records: {records}")
        print(f"  - Labeled: {labeled} ({labeled/records*100:.1f}%)")
        
        dataframes.append(df)
        total_records += records
    
    # Combine all dataframes
    if not dataframes:
        print("❌ No files found to merge. Please check the file paths.")
        return
    
    print("\n🔄 Combining all dataframes...")
    combined_df = pd.concat(dataframes, ignore_index=True)
    
    # Verify the column consistency
    print(f"\n📊 Combined dataset statistics:")
    print(f"  - Total records: {len(combined_df)}")
    
    # Check for sentiment label stats
    labeled_count = combined_df['sentiment_label'].notna().sum()
    print(f"  - Records with sentiment labels: {labeled_count} ({labeled_count/len(combined_df)*100:.1f}%)")
    
    # Check for duplicates
    duplicates = combined_df.duplicated(subset=['extracted_text']).sum()
    if duplicates > 0:
        print(f"⚠️ Found {duplicates} duplicate records. You may want to deduplicate the dataset.")
    
    # Count distribution of sentiment labels
    label_counts = combined_df['sentiment_label'].value_counts()
    print("\n📊 Sentiment label distribution:")
    for label, count in label_counts.items():
        print(f"  - {label}: {count} ({count/labeled_count*100:.1f}%)")
    
    # Save the combined dataset
    print(f"\n💾 Saving combined dataset to {output_file}...")
    combined_df.to_csv(output_file, index=False)
    print("✅ Merging completed successfully!")
    
    return combined_df

# Run the merging process
if __name__ == "__main__":
    merge_sentiment_files()
````
